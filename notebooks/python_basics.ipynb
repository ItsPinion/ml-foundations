# Python Basics


## print()
**print is a built-in function in Python that prints the value of the argument passed to it. It is used to display output on the ascreen.**



```python
print("Hello World!")
print("This is a simple " + "Python script with plus.")
print("This is a simple","Python script with comma.")
```

## vaiables
**A variable is a container that holds a value. we can get the value bu using the variable name.**


In python we dont need to declear the variable. We can directly use the variable name to print the value.

```python
x = 5
x = x + 10
print("x is", x)
```

## Datatypes

### Integers

it stores integers like 1, 2, 3 etc. All the number stored is whole numbers. 

```python
i = 5
print("i is", i)
```

### Floating point 

it stores a floating point number but the floating point number is only accurate to a certain number of digits.

```python
f = 3.1416
print("f is", f)
```

## Strings

string stores a sequence of characters or a sentence.



```python
s = "Hello World"
print(s)
```

We can use both single and double quotes to define strings.

```python
ss = "Hello World"
print(s)

sd = 'string in Python.'
print(sd)
```

We can also use Multiline strings to write long strings. 



```python
sm = """This is a string
that spans multiple lines."""
print(sm)
```

## Booleans

Booleans data type variables store only two kind of value, true/1 or false/0.

```python
b = True
print("Is the statement true?", b)
```

## Casting

All the variables in python have dynamic datatypes.

```python
x = 5
print(x)

x = "string"
print(x)

x = True
print(x)
```

But if we want to specify the datatypes we can use casting.

```python
i = int(5)
f = float(3.1416)
s = str("Hello World")
b = bool(True)

print(type(i))
print(type(f))
print(type(s))
print(type(b))
```
