### **Exercise 9: Longest Line**

- Read file line by line
- Find the line with the **maximum length**
- Print:
  - the line
  - its length


````python
file = open("../sample.txt", "r")

lines = file.readlines() 

if len(lines):
    
    longest = lines[0]

    for line in lines:
        if len(longest) < len(line):
            longest = line
    
    print(len(longest))
    print(longest)

file.close()
````

### **Exercise 11: Word Frequency (Single File)**

- Read file
- Build a dictionary:

  ```
  {
    "python": 1,
    "file": 2,
    ...
  }
  ```

- Case-insensitive
- Ignore punctuation if any


````python
file = open("../sample.txt", "r")

text = file.read().lower()

clean_text = ""

for char in text:
    if char.isalnum() or char.isspace():
        clean_text += char

words = clean_text.split()

frequency: dict[str, int] = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print(frequency)

file.close()
````

### **Exercise 12: Functionized Reader**

Write a function:

```python
def read_file(path: str) -> str:
    ...
```

- Returns full file content
- Does **not** print anything
- Caller prints result


````python
def read_file(path: str) -> str:
    file = open(path, "r")

    content = file.read() 
    
    file.close()

    return content
        
print(read_file("../sample.txt"))

````

### **Exercise 13: Safe Counter Function**

Write a function:

```python
def count_lines_and_words(path: str) -> tuple[int, int]:
    ...
```

Returns:

```
(lines, words)
```


````python
def count_lines_and_words(path: str) -> tuple[int, int]:
    file = open(path, "r")
    content = file.read()
    file.close()

    lines = content.splitlines()
    words = content.split()

    return len(lines), len(words)
````
