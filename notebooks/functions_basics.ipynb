### 4. Reverse String

Return a reversed version of a string.

```python

    pass
```

```python
reverse_string("hello")   # "olleh"
```

**Restriction:**

* Do NOT use slicing (`[::-1]`)

---

````python
def reverse_string(text:str)->str:
    reverse = ""
    for i in range(len(text)-1,-1,-1):
        reverse += text[i] 
    return reverse

print(reverse_string("hello"))
````


`count_even()` **must call `is_even()`**, not repeat logic.


---


````python
def is_even(n:int)->bool:
    
    return n % 2 == 0

def count_even(numbers:list[int])->int:
    count = 0
    for num in numbers:
        if is_even(num):
            count +=1
            
    return count

print(count_even([1, 2, 3, 4, 6]))
````


### 10. Average of Positives

Return the average of only positive numbers.

```python

    pass
```

Rules:

* Ignore 0 and negatives
* If no positive numbers, return `0`

```python
average_positive([1, -2, 3, 4])  # 2.666...
```

---

````python
def average_positive(numbers:list[int])->float:
    total = 0
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
            total += num
    if count == 0: return 0
    return total / count

print(average_positive([1, -2, 3, 4]))
````



### 12. Palindrome Check

Return `True` if string is palindrome.

```python
def is_palindrome(text):
    pass
```

Rules:

* Case-insensitive
* Ignore spaces

```python
is_palindrome("Madam")        # True
is_palindrome("nurses run")   # True
```

---



````python
def is_palindrome(text:str)->bool:
    
    fixed_text = text.replace(" ","").upper()

    for i in range(0,int(len(fixed_text)/2)):
        if fixed_text[i] != fixed_text[len(fixed_text) - i - 1]:
            return False
    
    return True


print(is_palindrome("Madam"))
print(is_palindrome("nurses run"))
````

## Exercise 1 – Temperature Converter

### Task

### Requirements

* Function name: `convert_temperature`
* Parameters:

### Test Cases You Must Try

* Celsius → Fahrenheit

* Call once using **keyword arguments** 

````python
def convert_temperature(value:float,unit:str)->float:
    if unit == "C":
        return (value *(9/5))+32
    elif unit == "F":
        return (value-32)* 5/9
    
    return 0

print(f"50 C = {convert_temperature(value=50,unit="C")} F")
print(f"50 F = {convert_temperature(value=50,unit="F")} C")
````

````python
def average(numbers:list[float])->float|None:
    count = len(numbers)
    if count == 0: return None
    
    total = 0
    
    for num in numbers:
        total += num
        
    return total/count

print(f"Avg: {average([547,456,9,241,58,2641,89,85,641])}")
print(f"Avg: {average([])}")
````


## Exercise 10 – Mini CLI Simulation

**Goal:** Argument-driven behavior

### Task

Create a function:

```python
def calculator(a, b, operation="add"):
```

### Behavior

* `"add"` → add
* `"sub"` → subtract
* `"mul"` → multiply
* `"div"` → safe division logic

Return result.



````python
def add(a:float,b:float)->float:
    return a + b

def sub(a:float,b:float)->float:
    return a - b

def mul(a:float,b:float)->float:
    return a * b

def div(a:float,b:float)->float|str:
    if b == 0:
        return ("Division by zero is not allowed.")
    return a / b

    
def calculator(a:float, b:float, operation:str="add")->float|str:
    if operation == "add":
        return add(a, b)
    elif operation == "sub":
        return sub(a, b)
    elif operation == "mul":
        return mul(a, b)
    elif operation == "div":
        return div(a, b)

    return "Invalid operation"

print(calculator(10, 5, "add"))
print(calculator(10, 5, "sub"))
print(calculator(10, 5, "mul"))
print(calculator(10, 5, "div"))
print(calculator(10, 0, "div"))
print(calculator(10, 5, "sqr"))
````
