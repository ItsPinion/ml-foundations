## Exercise 2
**Takes this list:**
```py
numbers = [1, 2, 3, 4, 5]
```
**Produces:**
```py
[5, 4, 3, 2, 1]
```
Rules:
- ❌ Do NOT use .reverse()

- ❌ Do NOT use slicing [::-1]

- ✅ Use a loop

- ✅ Build a new list

````python
numbers = [1, 2, 3, 4, 5]

reverse = []

for i in range(len(numbers) - 1,-1,-1):
    reverse.append(numbers[i])
    

print(reverse)
````

**Task**
Given a string:

```python
text = "programming"
```

Create a dictionary that counts how many times each character appears.

**Expected output (example)**

```python
{'p':1, 'r':2, 'o':1, 'g':2, 'a':1, 'm':2, 'i':1, 'n':1}
```

**Constraint**

* Do NOT use `collections.Counter`

---

````python
text = "programming"

temp = {}

for i in text:
    if(temp.get(i)):
        temp[i] += 1
    else:
        temp[i] = 1
        
print(temp)
````


## Section B — Dict vs List Thinking

### Exercise 4: Marks Analysis

Given:

```python
marks = {
    "Math": 85,
    "Physics": 78,
    "Chemistry": 90
}
```

**Tasks**

1. Print subject with highest marks
2. Calculate average marks
3. Add a new subject
4. Remove one subject

**Explain**
Why dict is better than list here (write comment).

---


````python
marks = {
    "Math": 85,
    "Physics": 78,
    "Chemistry": 90
}
highest = 0;
total = 0;
for i in marks.keys():
    total += marks[i]
    if marks[i] > highest:
        highest = marks[i]

print(highest)
print(f'avg {total/ len(marks.keys())}')

marks["Biology"]= 0
print(marks)

marks.popitem()
print(marks)
````

### Exercise 9: Word Grouping

Given:

```python
words = ["cat", "bat", "rat", "car", "bar"]
```

Group words by first letter.

**Expected**

```python
{
  'c': ['cat', 'car'],
  'b': ['bat', 'bar'],
  'r': ['rat']
}
```


````python
words = ["cat", "bat", "rat", "car", "bar"]

collection = {}

for word in words:
    collection[word[0]] = (collection.get(word[0]) or []) + [word]
    
print(collection)
````
