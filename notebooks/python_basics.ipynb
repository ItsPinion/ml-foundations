**Explain**

````python
marks = {
    "Math": 85,
    "Physics": 78,
    "Chemistry": 90
}
highest = 0;
total = 0;
for i in marks.keys():
    total += marks[i]
    if marks[i] > highest:
        highest = marks[i]

print(highest)
print(f'avg {total/ len(marks.keys())}')

marks["Biology"]= 0
print(marks)

marks.popitem()
print(marks)
````

**Goal:** Membership checking

**Task**

* Create a set called `visited_sites`
* Add websites one by one
* Before adding, check if the site was already visited

**Input Example**

```python
sites = ["google.com", "github.com", "google.com", "openai.com"]
```

**Output Logic**

* Print `"Already visited"` if duplicate
* Otherwise add to set




````python
visited_sites = set()
sites = ["google.com", "github.com", "google.com", "openai.com"]

for site in sites:
    if site in visited_sites:
        print("Already visited")
        continue
    visited_sites.add(site)
    
print(list(visited_sites))
````

registered = {"admin", "user1", "guest"}


## **Exercise 9: Words Appearing Once**

**Goal:** Combine loops + sets

**Task**

* Given a sentence
* Print words that appear only once

**Input**

```python
sentence = "python is easy and python is powerful"
```

**Hint**

* Use a set + loop
* You may also use a dictionary if needed


````python
sentence = ("python is easy and python is powerful").split(" ")
count_dic :dict[str, int]={}
unique = set()

for c in sentence:
    
    if(not count_dic.get(c)): count_dic.update({c:0})
    count_dic[c] += 1
    
for key, value in count_dic.items():
    if value == 1:
        unique.add(key)
   

print(unique)
````

**Goal:** Set growth logic

**Task**

* Iterate over a list

* Stop immediately when found

**Input**

```python
nums = [3, 1, 4, 2, 5, 1, 6]
```

**Hint**

* Maintain a `seen` set
* Check membership before adding


````python
nums = [3, 1, 4, 2, 5, 1, 6]
unique = set()
for i in nums:
    if i in unique:
        print(i)
        break
    unique.add(i)
````
